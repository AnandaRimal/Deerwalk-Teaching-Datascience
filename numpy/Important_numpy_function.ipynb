### Summary of `axis` in NumPy

#### General Rule:
- **`axis=0`**: Refers to **rows** (vertical direction).
- **`axis=1`**: Refers to **columns** (horizontal direction).

#### Behavior by Operation:
1. **Aggregation Operations (e.g., `sum`, `mean`, `min`, `max`)**:
   - **`axis=0`**: Collapses rows (performs operation **column-wise**).
   - **`axis=1`**: Collapses columns (performs operation **row-wise**).

2. **Sorting Operations (e.g., `np.sort`)**:
   - **`axis=0`**: Sorts **column-wise** (each column sorted independently).
   - **`axis=1`**: Sorts **row-wise** (each row sorted independently).

3. **Concatenation Operations (e.g., `np.append`, `np.concatenate`)**:
   - **`axis=0`**: Stacks arrays **vertically** (adds rows).
   - **`axis=1`**: Stacks arrays **horizontally** (adds columns).

#### Key Takeaways:
- **`axis=0`**:
  - In aggregation/sorting: Operates **column-wise** (across rows).
  - In concatenation: Adds **rows** (vertical stacking).
- **`axis=1`**:
  - In aggregation/sorting: Operates **row-wise** (across columns).
  - In concatenation: Adds **columns** (horizontal stacking).

## np.concatenate() ##

The np.concatenate() function is used to join two or more arrays along an existing axis. You can specify the axis along which you want to concatenate the arrays using the axis parameter.

- **`np.concatenate()`**: Joins arrays along an existing axis (you specify which axis).
- **`np.vstack()`**: Stacks arrays vertically (row-wise), equivalent to `np.concatenate()` along `axis=0`.
- **`np.hstack()`**: Stacks arrays horizontally (column-wise), equivalent to `np.concatenate()` along `axis=1`.

using concatenate we can avoid using vstack() and hstack()

````python
a = np.array([[1,2],[3,4]])
a
````

````python
b = np.array([[5,6]])
b
````

````python
# now concatenate both
np.concatenate((a,b),axis=0) #add rows(like in append)
````

````python
np.concatenate((a,b),axis=1) #add columns(like in append)
````

````python
# b should be reshaped or transpose
b = np.transpose(b)
np.concatenate((a,b),axis=1)
````

## Using `np.expand_dims()` in Machine Learning

In machine learning, models usually expect inputs in batches. When making predictions, a model takes input and provides output. For instance, if we have a model that accepts goals/assists and age and predicts a footballer's market value, most machine learning algorithms expect multiple inputs at the same time, not just data for one player.

If you have data for a single player with shape `(2,)` (letâ€™s say goals/assists and age), this shape means you only have one example. If your model expects batch inputs, you need to transform it to include a batch dimension. Using `np.expand_dims()` can add this dimension, changing the shape from `(2,)` to `(1, 2)`. Now, it looks like you have a batch of one player data point.

Here's how it works in practice:

```python
import numpy as np

# Single player data: [g/a, age]
player_data = np.array([30, 25])

# Expand dimensions to simulate a batch of one
player_data_batch = np.expand_dims(player_data, axis=0)

print(player_data_batch.shape)  # Output: (1, 2)


## np.where() ##

The `np.where()` function in NumPy is a versatile tool for conditional selection and manipulation of array elements. It allows for the evaluation of a condition and provides options for choosing values based on whether the condition is `True` or `False`.

## Syntax:
```python
np.where(condition, [x, y])
condition: A boolean expression or an array-like structure that evaluates to True or False for each element in an array.
x (optional): An array or scalar specifying the values to be returned where the condition is True.
y (optional): An array or scalar specifying the values to be returned where the condition is False.


````python
# lets do the example of np.where() which is one of the very useful function
a
````

````python
# now find all the indices with value greater than 40
np.where(a>40)
````

````python
#replace all the values > 40 with 0
np.where(a>40,0,a) #return 0 where condition is true and a(existing value from array a ) if condition is false
````

````python
# replace all even value with o
np.where(a%2 == 0,0,a)
````

Q. Given an array of integers, use np.where() to create a new array that labels each element as 'Even' or 'Odd'.

````python
arr = np.array([1,2,3,4,5,6])
np.where(arr%2 == 0 ,"Even","Odd")
````

Q. Replace Negative Values: Given an array of integers, use np.where() to replace all negative values with zero.

````python
arr = np.array([-5, 3, -1, 7, -2, 8])
np.where(arr < 0,0,arr)
````

Q. Conditional Transformation: Given two arrays of the same shape, use np.where() to create a new array that takes elements from the first array if they are positive and from the second array if they are non-positive.

````python
arr1 = np.array([1, -2, 3, -4, 5])
arr2 = np.array([-1, 2, -3, 4, -5])
np.where(arr1>0,arr1,arr2)
````

## np.cumsum()
np.cumsum() function is used when we want to compute the cumulative sum of array elements over a given axis .
keep in mind that it is not like sum rather it returns array with cummulative sum (summing with previous value)
for example we have a goal of cristiano ronaldo in match as [2,3,1,5,1,0,0,2,1].
then the cummulative sum is like [2,2+3,2+3+1,....]

````python
#example 
ronaldo_last10_goals = ronaldo_goals = np.array([1, 2, 3, 1, 0, 0, 2, 1, 1, 0])
ronaldo_last10_goals_cummulative = np.cumsum(ronaldo_last10_goals)
ronaldo_last10_goals_cummulative
````

using this we can see the total goals of ronaldo at any point like after 8th game total goal was 10. This is very useful function

In 2d arrays, if axis is not defined then it will convert 2d array to 1d , we should specify axis to find the cummulative sum of that axis.
