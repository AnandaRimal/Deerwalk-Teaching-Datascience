```python
# Polynomial Regression Notebook

# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

```

```python
 # Step 2: Create Sample Dataset
# Example: Relationship between x and y is non-linear: y = x^2 + noise
np.random.seed(0)
X = 2 - 3 * np.random.rand(20, 1)
y = X**2 + np.random.randn(20, 1)
```

```python

# Visualize the dataset
plt.scatter(X, y, color='blue')
plt.title('Sample Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.show()

```

```python
# Step 3: Linear Regression (for comparison)
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_lin = lin_reg.predict(X)

plt.scatter(X, y, color='blue')
plt.plot(X, y_pred_lin, color='red', label='Linear Regression')
plt.title('Linear Regression vs Data')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

```

```python
# Step 4: Polynomial Regression
# Transform features to polynomial features (degree = 2)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)

# Train Linear Regression on transformed features
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
y_pred_poly = poly_reg.predict(X_poly)

```

```python
# Visualize Polynomial Regression
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred_poly, color='green', label='Polynomial Regression')
plt.title('Polynomial Regression vs Data')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()

```

```python
# Step 5: Evaluate Models
print("Linear Regression R^2:", r2_score(y, y_pred_lin))
print("Polynomial Regression R^2:", r2_score(y, y_pred_poly))

```

```python
print("Linear Regression MSE:", mean_squared_error(y, y_pred_lin))
print("Polynomial Regression MSE:", mean_squared_error(y, y_pred_poly))

```

```python
# Step 6: Predict new values
X_new = np.array([[1.5], [3.0], [-1.0]])
X_new_poly = poly_features.transform(X_new)
y_new_pred = poly_reg.predict(X_new_poly)

print("Predictions for new data (Polynomial Regression):")
for i, x_val in enumerate(X_new):
    print(f"X={x_val[0]}, Predicted y={y_new_pred[i][0]:.2f}")
```
