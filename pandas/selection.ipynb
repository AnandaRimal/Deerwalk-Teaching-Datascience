```python
import pandas as pd

df = pd.read_csv("datasets/Titanic-Dataset.csv")
df.head()
```

```python
#Single column → Series
df["Age"]

```

```python
df[["Name", "Sex", "Age"]]
```

```python
### .loc — Label-based Selection (VERY IMPORTANT)
df.loc[row_label, column_label]
```

```python
df.loc[0:4]
```

```python
df.loc[0:4, ["Name", "Age", "Sex"]] #Select rows + columns
```

```python
df.loc[0, ["Name","Age"]] #Select a single value (cell)
```

```python
# Select all rows but specific columns
df.loc[:, ["Survived", "Pclass", "Fare"]]
```

### .iloc — Position-based Selection

df.iloc[row_position, column_position]


```python
df.iloc[0:5] #Select first 5 rows

```

```python
df
```

```python
df.iloc[0:5, 0:3] # Select rows & columns by position
```
