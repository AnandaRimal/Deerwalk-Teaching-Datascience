```python
#Passengers in 1st class
df.loc[df["Pclass"]==1,["Name","Age","Pclass","Sex"]]
```

```python
# Age > 30 AND Survived
df.loc[(df["Age"] > 30) & (df["Survived"] == 0), :]
```

```python
# Age < 18 OR Female
df.loc[(df["Age"] < 18) | (df["Sex"] == "female"), :]
```
