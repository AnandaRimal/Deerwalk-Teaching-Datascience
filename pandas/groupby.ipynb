3. What is GroupBy?

groupby() is used to split data into groups, apply a function, and combine results.

Split → Apply → Combine

Basic GroupBy (Single Column)
Group by Department

```python
df.groupby('Department')
```

```python
df.groupby('Department')['Salary'].mean()
#Calculate Mean Salary per Department
```

```python
# Multiple Aggregations
df.groupby('Department')['Salary'].agg(['mean', 'max', 'min', 'sum'])
```

```python
# GroupBy with Multiple Columns 
df.groupby(['Department', 'Experience'])['Salary'].mean()
```

```python
df.groupby('Department')['Employee'].count()
#Using GroupBy with Count
```

```python
def salary_range(x):
 return x.max() - x.min()


df.groupby('Department')['Salary'].apply(salary_range)
```

```python
# Test with new data
import pandas as pd

students = pd.DataFrame({
    'student_id': [101, 102, 103, 104],
    'student_name': ['Ram', 'Sita', 'Hari', 'Gita'],
    'department': ['CS', 'IT', 'CS', 'BCA']
})

print(students)
```

```python
marks = pd.DataFrame({
    'student_id': [101, 102, 105],
    'marks': [88, 92, 75]
})

print(marks)
```

```python
inner_join = pd.merge(students, marks, on='student_id', how='inner')
print(inner_join)
# Common records only
```

```python
left_join = pd.merge(students, marks, on='student_id', how='left')
print(left_join)
```

```python
right_join = pd.merge(students, marks, on='student_id', how='right')
print(right_join)
```

```python
outer_join = pd.merge(students, marks, on='student_id', how='outer')
print(outer_join)
# Outer Join (Everything from both tables)
```

Easy Memory Trick

Inner → common only

Left → everything from first table

Right → everything from second table

Outer → everything from both
