```python
import pandas as pd

data = {
    "Name": ["Ram", "Sita", "Hari", "Gita"],
    "Age": [22, 19, 25, 21],
    "Marks": [78, 85, 67, 90]
}

df = pd.DataFrame(data)
df
```

```python
#Sort by Age
df.sort_values(by="Age")
```

```python
# Sort by Marks (descending)
df.sort_values(by="Marks", ascending=False)
```

```python
# First by Age, then by Marks
df.sort_values(by=["Age", "Marks"])
```

```python
df.sort_values(by=["Age", "Marks"], ascending=[True, False])
# Different order for each column:
```

```python
df.sort_index()
# sort by index
```

```python
df.sort_index(ascending=False)
```

```python
# Sort and save changes (inplace)
df.sort_values(by="Marks", inplace=True)
```

```python
# Sort strings (alphabetical order)
df.sort_values(by="Name")
```

```python
# Reverse alphabetical:
df.sort_values(by="Name", ascending=False)
```

## ðŸ§  Quick Summary

| Task | Method |
|------|--------|
| Sort rows | `sort_values()` |
| Sort index | `sort_index()` |
| Descending order | `ascending=False` |
| Multiple columns | `by=[...]` |
| Save changes | `inplace=True` |
| Fix index | `reset_index(drop=True)` |
