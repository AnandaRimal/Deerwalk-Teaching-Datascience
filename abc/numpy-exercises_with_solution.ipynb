```python
# Create two arrays of random integers between 0 to 10
# both of size (4, 3)
rand_arr4 = np.random.randint(0,11,(4,3))
rand_arr5 = np.random.randint(0,11,(4,3))
print(rand_arr4)
print(rand_arr5)
```

```python
# Perform a dot product on the two newest arrays you created
dot2 = np.dot(rand_arr4,rand_arr5)
dot2
```

```python
#first we have to reshape one of the array
dot2 = np.dot(rand_arr4.reshape(3,4),rand_arr5)
dot2
```

It doesn't work. How would you fix it?

```python
# Take the latest two arrays, perform a transpose on one of them and then perform 
# a dot product on them both
dot3 = np.dot(np.transpose(rand_arr4),rand_arr5)
dot3
```

Notice how performing a transpose allows the dot product to happen.

Why is this?
